Visualizing our data.

```python
df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
my_cols = ["Insulin", "BMI", "Glucose", "BloodPressure", "SkinThickness"]
```

```python
for col in my_cols:
    plt.figure(figsize=(12,8))
    sns.boxplot(x="Outcome", y=col, data=df)
    plt.title(f"Boxplot of {col} by Outcome")
    plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Corr Heatmap")
plt.show()
```

```python
df["Glucose_Insulin"] = df["Glucose"] * df["Insulin"]
```

```python
standard_scaler = StandardScaler()
scaled = pd.DataFrame(standard_scaler.fit_transform(df), columns=df.columns)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(df["Glucose"], kde=True)
plt.title("Histogram / Glucose")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(y="BloodPressure", data=df)
plt.title("Boxplot / BloodPressure")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="Age", y="Glucose", hue="Outcome", data=df)
plt.title("Scatter Plot Glucose vs Age")
plt.show()
```

```python
subset = df[["Age", "Glucose", "BloodPressure", "BMI"]]
plt.figure(figsize=(12,8))
sns.heatmap(subset.corr(), annot=True, cmap="coolwarm")
plt.title("Corr Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x="Outcome", y="Glucose", data=df)
plt.title("Bar Plot Glucose")
plt.show()
```
