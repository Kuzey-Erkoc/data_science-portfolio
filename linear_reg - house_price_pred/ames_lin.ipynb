Getting started with machine learning and using GridSearchCV and Cross Validation.

```python
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
base_elastic_model = ElasticNet(max_iter=1000)
```

```python
param_grid = {"alpha": [0.1, 1, 5, 10, 100], "l1_ratio": [.1, .7, .99, 1]}
```

```python
grid_model = GridSearchCV(base_elastic_model, 
                          param_grid=param_grid,
                          scoring="neg_mean_squared_error",
                          cv=5, verbose=1)
```

```python
grid_model.fit(scaled_X_train, y_train)
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
